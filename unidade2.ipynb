```python
# Atividade 1)

import pandas as pd

# Criando o DataFrame com os dados fornecidos
dados = {
    "Data": ["15/01/2025"] * 5,
    "Cidade": ["São Paulo", "Rio de Janeiro", "Curitiba", "Porto Alegre", "Salvador"],
    "Temperatura Máxima (°C)": [30.5, 35.0, 24.0, 28.0, 31.0],
    "Temperatura Mínima (°C)": [22.0, 25.0, 18.0, 20.0, 24.5],
    "Precipitação (mm)": [12.0, None, 8.0, 15.0, None],
    "Umidade Relativa (%)": [78, 70, None, 82, 80]
}

df = pd.DataFrame(dados)

# Substituindo valores ausentes
df["Precipitação (mm)"] = df["Precipitação (mm)"].fillna(df["Precipitação (mm)"].mean())
df["Umidade Relativa (%)"] = df["Umidade Relativa (%)"].fillna(df["Umidade Relativa (%)"].median())

# Criando a coluna Amplitude Térmica
df["Amplitude Térmica"] = df["Temperatura Máxima (°C)"] - df["Temperatura Mínima (°C)"]

# Filtrando cidades com Temperatura Máxima acima de 30°C
df_cidades_quentes = df[df["Temperatura Máxima (°C)"] > 30]

# Reordenando as colunas
df = df[["Data", "Cidade", "Temperatura Máxima (°C)", "Temperatura Mínima (°C)",
         "Amplitude Térmica", "Precipitação (mm)", "Umidade Relativa (%)"]]

# Exibindo os resultados
print("DataFrame Completo:")
print(df)

print("\nDataFrame Cidades com Temperatura Máxima acima de 30°C:")
print(df_cidades_quentes)
```

```python
# Atividade 2)

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Dados fictícios
dados_vendas = pd.DataFrame({
    "Dia da Semana": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"],
    "Vendas Totais ($)": [1200, 1500, 1300, 1700, 1800, 2200, 1600],
    "Número de Clientes": [50, 60, 55, 70, 75, 90, 65],
    "Lucro ($)": [300, 400, 350, 500, 600, 700, 450]
})

# 1. Gráfico de Barras - Total de vendas por dia
plt.figure(figsize=(8, 5))
sns.barplot(x="Dia da Semana", y="Vendas Totais ($)", data=dados_vendas)
plt.title("Total de Vendas por Dia")
plt.xlabel("Dia da Semana")
plt.ylabel("Vendas Totais ($)")
plt.show()

# 2. Gráfico de Dispersão - Número de clientes vs Vendas Totais
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Número de Clientes", y="Vendas Totais ($)", data=dados_vendas)
plt.title("Número de Clientes vs Vendas Totais")
plt.xlabel("Número de Clientes")
plt.ylabel("Vendas Totais ($)")
plt.show()

# 3. Heatmap - Correlação entre as variáveis
corr = dados_vendas[["Vendas Totais ($)", "Número de Clientes", "Lucro ($)"]].corr()

plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlação entre Vendas, Número de Clientes e Lucro")
plt.show()
```

```python
# Atividade 3)

import matplotlib.pyplot as plt

# Definindo os horários (eixo X)
horarios = list(range(25))  # De 0h a 24h

# Definindo a temperatura (eixo Y)
temperaturas = [15, 16, 17, 18, 19, 20, 22, 24, 26, 28, 29, 30, 30, 29, 28, 26, 24, 22, 21, 20, 19, 18, 17, 16, 15]

# Criando o gráfico
plt.figure(figsize=(10, 5))  # Ajustando o tamanho

plt.plot(horarios, temperaturas, marker="o", linestyle="-", color="royalblue", markersize=6, label="Temperatura")

# Personalizações básicas
plt.title("Evolução da Temperatura Durante o Dia", fontsize=14, fontweight="bold")
plt.xlabel("Horário (h)", fontsize=12)
plt.ylabel("Temperatura (°C)", fontsize=12)

# Ajustando os ticks dos eixos
plt.xticks(range(0, 25, 2))  # De 2 em 2 horas
plt.yticks(range(15, 32, 2)) # Intervalos de temperatura

# Adicionando uma grade leve
plt.grid(True, linestyle="--", alpha=0.5)

# Adicionando legenda
plt.legend()

# Exibir o gráfico
plt.show()
```
